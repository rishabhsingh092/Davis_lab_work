```python
# Function to check palindrome
def check_palindrome(value):
    value = str(value)  # Convert input to string
    reversed_value = ""

    # Use a for loop to reverse the string
    for char in value:
        reversed_value = char + reversed_value

    # Check if original and reversed are the same
    if value == reversed_value:
        print(value, "is a Palindrome")
    else:
        print(value, "is NOT a Palindrome")

# Input from user
user_input = input("Enter a number or string: ")
check_palindrome(user_input)
```
