```python
#Salry_bonus_calculator

def bonus(salary, exp):
    if exp >= 10:
        return salary * 0.20
    elif exp >= 5:
        return salary * 0.10
    else:
        return salary * 0.05

for i in range(3):
    sal = float(input("Enter salary: "))
    exp = int(input("Enter experience: "))

    print("Bonus:", bonus(sal, exp))
```
