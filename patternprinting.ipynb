```python
# Input number of rows
rows = int(input("Enter number of rows: "))

# Loop through each row
for i in range(1, rows + 1):
    # Check if row is even or odd
    if i % 2 == 0:
        symbol = "*"  # even row
    else:
        symbol = "#"  # odd row

    # Print symbols for the row
    for j in range(i):
        print(symbol, end=" ")
    print()  # move to next line
```
